**ANS 1**


**Overfitting:** Overfitting occurs when a machine learning model learns the training data too well, capturing noise or random fluctuations that don;t represntthe true underlying patterns. Consequences include poor generalization to new data, high accuracy on the training set but low accuracy on the test set.

**Underfitting:** Underfitting happens when a model is too simple to captutre underlying patterns in training data. Consequences include low accuracy on both the training and test sets, indicating that the model has not learned the underlying structure.

Mitigation:



**Overfitting:** Regularization techniques, cross-validation, reducing model complexity, and increasing the amount of training data can help mitigate overfitting.

**Underfitting:** Increasing model complexity, using a more sophisticated algorithm, or providing more relevant features can address underfitting.

**ANS 2**


To reduce overfitting:

1. Regularization: Introduce penalties for complexities in the model discouraging overly intricate patterns.

2. Cross-validation: Evaluate the model's performance on different subsets of the data to ensure generalization.

3. Pruning: In decision trees, remove unnecessary branches that capture noise.

4. Increasing Data: Provide more diverse and representative data for training


**ANS 3**
Underfitting: Underfitting occurs when a model is too simple to capture underlying patterns in the data.

Scenarios where underfitting can occur:

1. Using a linear model for a highly non-linear problem.

2. Applying a simple algorithm to a complex dataset.

3. Insufficient training data for a more complex model.

**ANS 4**

**BIAS-VARIANCE TRADEOFF** : It's a balancing act between bias (error from overly simplistic assumptions ) and variance (Sensitivty to fluctuations in the training set). High bias leads to underfitting, High variance leads to overfitting.


**Relationships**: As you decrease bias, variance tends to increase and vice versa.


**Effect on Performance**: Finding the right balance is crucial. Models with an optimcal bias-variance tradeoff generalize well to new data.

**Overfitting Detection:** 

1. Evaluate on a separate test set.
2. Monitor the performancce metrics on both training and test sets.
3. Use learning curves to visualize the training and validation performance.
4. Check for excessively high accuracy on the training set but lower accuracy on the test set.

**Underfitting Detection:**

1. Similar approaches apply, focusing on performance metrics and learning curves.
2. If the model performs poorly on both training and test sets, it might be underfitting.

**ANS 7**

Regularization: Regularization introduces a penalty term to the model's cost function to discourage complex models and prevent overfitting.

Common Techniques:

1. L1 Regularization(Lasso): Adds the absolute values of the coefficients to the cost function.

2. L2 Regularization(Ridge): Adds the squared values of the coefficients to the cost function.

3. Dropout: Randomly drops some neurons during training in neural networks, preventing over-reliance on specific neurons.

4. Early Stopping: Halts training when performance on a validation set starts to degrade.
