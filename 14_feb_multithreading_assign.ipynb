**ANS 1**

Multithreading is a concurrent execution technique in Python that allows multiple threads to run in a single process. A thread is a lightweight sub-process that shared the same memory space as the parent process but executes independently. Threads are used to achieve parallelism, allowing multiple tasks to be executed simultaneously.


**ANS 2**


The module used to handle threads in Python is the "threading" module. 

    1. activeCount() : This function returns the number of thread objects concurrently alive.
    2. currentThread(): This function returns the current Thread object, corresponding to the caller's thread of control.
    3. enumerate(): This function returns a list of all Thread objects currently alive.

    


**ANS 3**

1. run(): This method is executed when a thread is started. It contains the code that the thread will execute.
2. start(): This method is used to start a thread's execution. It automatically calls the "run" method.
3. join(): This method blocks the calling thread until the thread it is called on completes its execution.
4. isAlive(): This method returns "True" if the thread is currently executing otherwise "False".

**ANS 4**



```python
import threading
def print_squares():
    for i in range(1,11):
        print(f'Square of {i}: {i**2}')
def print_cubes():
    for i in range(1,11):
        print(f'Cube of {i}: {i**3}')
thread1=threading.Thread(target=print_squares)
thread2=threading.Thread(target=print_cubes)
thread1.start()
thread2.start()
thread1.join()
thread2.join()

print('both threads have finished')
```

ADVANTAGES:
1. Improved Perfomance by utilizing available CPU cores for parallel execution.
2. Responsiveness: Allows programs to remain responsive even when performing resourse-intensive tasks.
3. Simplified program structure for tasks that can be executed concurrently.
4. Efficient utilization of system resources.

DISADVANTAGES:
1. Increased complexity due to synchronization and potential race conditions.
2. Debugging multithread programs can be challenging.
3. Overhead due to thread creation and management.
4. GIL (Global Interpreter Lock) In CPython limits true parallelism in CPU-bound tasks.


**ANS 6**

Deadlocks: A deadlock occurs when two or more threads are unable to procees because each is waiting for a resource held by another. This can result in a situation where all threads are blocked and the program comes to a standstill.

Race Conditions: A race condition occurs when multiple threads access shared data concurrently, and at least one of them modifies the data. The output of the program depends on the order in which the threads execute, leading to the unpredictable behaviour.


Both deadlocks and race conditions are common concurrency issues that need to be carefully managed in multithreaded programming.
