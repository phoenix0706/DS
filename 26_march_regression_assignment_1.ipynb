**ANS 1**


Simple Linear Regression: In simple linear regression, a single independent variable (predictor variable) is used to predict a dependent variable. The relationship between the variables is assumed to be a straight line. An example could be predicting a person's weight based on their height.



Multiple Linear Regression: Multiple linear regression involves multiple independent variables to predict a dependent variable. It assumes a linear relationship, but in this case, there are multiple predictors involved. For instance, predicting a house's price based on its square footage, number of bedrooms, and location.



**ANS 2**

Assumptions for Linear Regression are:

1. Linearity: The relationship between predictors and the response is linear.

2. Independence: Residuals (the differences between actual and predicted values are independent of each other)

3. Homoscedasticity: Residuals have constant variance across all levels of predictors.

4. Normality: Residuals are normally distributed.

5. No or Low Multicollinearity: Predictors are not highly correlated with each other. 

To check these assumptions, we can analyze residual plots,
perform normality tests on residuals, check for multicollinearity using correlation matrices or variance inflation factors, and so on.

**ANS 3**

In a Linear Regression equation, the slope represents the change in the dependent variable for a one-unit change in the independent variable, assuming other variables are held constant. 

The intercept is the value of dependent variable when all independent variables are 0.


Example: Suppose you're analyzing the relationship between years of experience and salary. In the equation Salary = 3000 + 1000 * Experience, the intercept is $3000, indicating the expected salary for someone with zero experience. The slope of 1000 suggests that for each additional year of experience, the salary is expected to increase by $1000.

Gradient descent is an optimization technique used in machine learning to find optimal parameters (weights and biases) of a model that minimizes a loss function. It works by iteratively adjusting the parameters in the direction of steepest descent of the loss function, which is calculated using gradients. This process continues until the algorithm converges to a minimum, hopefully the global minimum.

**ANS 5**

Multiple linear regression extends simple linear regression to include multiple predictors. The model equation is:
Y=B0+B1.X1+B2.X2+ ... +Bn.Xn+E
here Y is the dependent variable, X1, X2,.. Xn are independent variables and B0,B1, B2,...Bn are the coefficients and the E is the error term

**ANS 6**


Multicollinearity occurs when two or more independent variables in a multiple regression model are highly correlated. This can make it difficult to isolate the individuals effects of the predictors on the dependent variable.


To detect multicollinearity, you can calculate correlation matrices between predictors and use variance inflation factors (VIFs). To address it, you might consider removing one of the correlated variables or using dimensionality reduction techniques.

**ANS 7**


Polynomial Regression is an extension of linear regression where the relationship between the independent and dependent variables is modeled as an nth-degree polynomial. This allows the model to capture more complex relationships that can't be adequately represented by a straight line.

**ANS 8**


Advantages:

1. Can capture non-linear relationships.
2. More flexible and can fit a wider range of data patterns.
3. Allows better representation of data with curvatures.



Disadvantages:

1. Prone to overfitting, especially with higher-degree polynomials. 
2. Interpretability can be challenging with complex polynomial mode.s
3. Extrapolation outside the range of observed data can be unreliable.

Use polynomial regression when you have domain knowledge suggesting a non-linear relationship and when simple linear regression doesn't fit the data well. Be cautious with higher-degree polynomials to avoid overfitting.
