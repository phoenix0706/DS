**ANS 1**



The three measures of central tendency are:

Mean: The mean, also known as the average, is calculated by adding up all the values in a dataset and then dividing by the total number of values. It is sensitive to extreme values, as a single outlier can greatly affect the mean.



Median: The median is the middle value in a dataset when it is ordered from least to greatest. If the dataset has an odd number of values, the median is the middle value. If the dataset has an even number of values, the median is the average of the two middle values. The median is less affected by outliers compared to the mean.



Mode: The mode is the value that appears most frequently in a dataset. A dataset can have no mode (when all values occur with the same frequency), one mode (when a single value occurs most frequently), or multiple modes (when multiple values have the same highest frequency). The mode is useful for categorical data or when looking for the most common value.

Measures of Dispersion and Their Use:


Range: The range is the difference between the maximum and minimum values in a dataset. It gives a rough idea of how spread out the data is.


Variance: Variance measures the average of the squared differences from the mean. It quantifies how much individual data points deviate from the mean.


Standard Deviation: The standard deviation is the square root of the variance. It provides a measure of the average amount of deviation of data points from the mean.


Example: If you have two datasets, A and B, with the same mean but different standard deviations, the dataset with the higher standard deviation is more spread out, indicating greater variability in the data points.

Covariance: Covariance measures the degree to which two variables change together. It indicates the direction of the linear relationship between variables but doesn't provide a standardized measure.


Correlation: Correlation is a standardized measure of the strength and direction of the linear relationship between two variables. It ranges from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no linear correlation.


These measures are used in statistical analysis to understand relationships between variables and to assess the degree of association between them.

**ANS 14**

Outliers can significantly affect measures of central tendency and dispersion:

Mean: Outliers can pull the mean in their direction, making it an unreliable representation of the central value.

Median: The median is less affected by outliers because it's based on the position of values rather than their magnitude.

Mode: Outliers might introduce new modes or alter the mode, but their impact is usually smaller than on mean.


Range, Variance, and Standard Deviation: Outliers can increase the range, variance, and standard deviation due to their large deviations from the rest of the data.


Example: In a dataset of exam scores where most students score between 70 and 80, an outlier score of 150 can drastically increase the mean and the standard deviation,
 even though it doesn't represent the typical student performance.
