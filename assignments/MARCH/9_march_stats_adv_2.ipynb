Probability Mass Function(PMF): The PMF is a function that gives the probability of a discrete random variable taking a specific value. It defines the probabilities of all possible outcomes of a discrete random variable.


Example: Consider rolling a fair six-sided dice. The PMF of the die's outcomes would be: P(X=1)=1/6, P(X=2)=1/6, ... , P(X=6)=1/6, where X is the random variable representing the outcome.

Probability Density Function(PDF): The PDF is a function that describes the probability distribution of a continuous random variable. Unlike discrete variables, continuous varibales can take any value within a certain range, making it impossible to assign a specific probability to individual points.

Example: Let's say we have a random variable representing the height of individuals in a population. The PDF would describe how the height values are distributed across the range of possible heights.

Normal Distribution, also known as the Gaussian Distribution, is a symmetric bell-shaped probability distribution that is commonly used to model various real-world phenomena. Some examples of situattions where the normal distributon might be used include:

1. Height: Human height often follows a normal distribution pattern

2. IQ Scores: Intelligence quotient scores tend to be normally distributed.

3. Errors in Measurement: In many scientific measurements, errors follow a normal distribution.

4. Financial Returns: Stock market returns are often modeled using the normal distribution. 

5. Natural Phenomena: Some natural phenomena, like the distribution of particle velocities in a gas, can be approximated using normal distribution

The parameters of the normal distribution are "mean" and "standard deviation". 

The mean determines the central location of the distribution, while the standard deviation controls the spread or dispersion of the data. A larger standard deviation leads to a wider distribution, and a smaller standard deviation results in a narrower distribution.

1. Statistical Inference: Many statistical methods assume normality, making the normal distribution a foundational concept in statistics.

Bernoulli Distribution: The Bernoulli distribution models a binary outcome where an event  can have only two possible outcomes, usually denoted as 0 and 1. It is characterized by a single parameter p, representing the probability of the "success" outcome (usually denoted as 1)

Example: Tossing a coin (heads/tails). Let's say "success" (1) represents getting heads and "failure" (0) represents getting tails. The probability of heads (success) is p, and the probability of tails (failure) is 1 - p.

Binomial DisributionL The binomial distribution models the number of "successes" in a fixed number of independent Bernoulli Trials. It is characterized by two parameters: the number of trials (n) and the probability of success (p).

Difference: The key difference between the Bernoulli and binomial distribution is that the Bernoulli distribution deals with a single trial, while the binomial distribution deals with multiple independent trials.

**ANS 6**

The probability that a randomly selected observation will be greater than 60  is given below

```python
import scipy.stats as stats

# Parameters
mean = 50
std_dev = 10
value = 60

# Calculate z-score
z_score = (value - mean) / std_dev
print("Z-score: ",z_score)

# Calculate the probability using the cumulative distribution function (CDF)
probability = 1 - stats.norm.cdf(z_score)

print("Probability:", probability)
```

The Central Limit Theorem states that the distribution of the sample means approaches a normal distribution as the sample size increases, regardless of the original distribution of the population, as long as the sample size is sufficiently large. This theorem is a fundamental concept in statistics and has wide-ranging implications for hypothesis testing and confidence interval estimation.

Significance: The CLT enables us to make inferences about population parameters based on sample data. It provides a way to work with the normal distribution even when the underlying data may not be normally distributed.



**ANS 10**

The Central Limit Theorem relies on a few assumptions:

Independence: The individual observations in the sample must be independent of each other.
Sample Size: The larger the sample size, the better the approximation to a normal distribution.
Finite Variance: The population should have a finite variance (i.e., not be extremely skewed or have heavy tails).

Even if these assumptions are only partially met, the Central Limit Theorem tends to hold well for practical purposes, especially as the sample size becomes larger.
