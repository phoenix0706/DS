Probability Density function (PDF) is a concept of probability theory and statistics. It is used to describe the likelihood of a continuous random variable taking on a specific value. In contrast to discrete probability distributions, where probabilities are assigned to individual points, the PDF assigns probabilities to intervals. The integral of PDF over an interval gives the probability that the random variable falls within that interval. 

For a continuous random variable X, the PDF is denoted as f(x) and satisfies the property that the area under the curve within any interval is equal to the probability of X falling within that interval.

Below is th python function to calculate pdf of normal distibution with given mean and standard deviation at a given point


```python
import math
def normal_pdf(x,mean,std_dev):
    exponent=-((x-mean)**2)/(2*std_dev**2)
    coefficient=1/(std_dev*math.sqrt(2*math.pi))
    pdf_value=coefficient*math.exp(exponent)
    return pdf_value
mean=0
std_dev=1
x_value=1.5
pdf_at_x=normal_pdf(x_value,mean,std_dev)
print("PDF at x=",x_value,":",pdf_at_x)
```

**ANS 4**

Properties of Binomial Distribution are:

1. It consists of a fixed number of independent trials
2. Each trial has only two possible outcomes: success or failure
3. The probability of success remains constant across all trials
4. The trials are independent of each other


Examples of events modeled by Binomial Distribution:

1. Flipping a coin a certain number of times and counting the number of heads.
2. Conducting a series of medical tests on a group of patients and recording the number of patients who test positive.

```python
import numpy as np
import matplotlib.pyplot as plt
sample_size=1000
success_prob=0.4
# Generate random sample from binomial distribution
sample=np.random.binomial(1,success_prob,sample_size)
# Plot histogram
plt.hist(sample,bins=2,align='mid',rwidth=0.8,color='blue',alpha=0.7)
plt.xlabel('Sucess(1)/Failure(0)')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution')
plt.show()
```

```python
import math
def poisson_cdf(x,mean):
    cdf=0
    for k in range(x+1):
        cdf+=(math.exp(-mean)*mean**k)/math.factorial(k)
    return cdf
mean=3
x_value=5
cdf_at_x=poisson_cdf(x_value,mean)
print("CDF at x =", x_value, ":", cdf_at_x)
```

**ANS 7**

Binomial Distribution:

1. Describes the number of successes in a fixed number of independent trials.
2. Each trial has two possible outcomes: success or failure.
3. The probability of success remains constant across trials.
4. Discrete distribution.


Poisson Distribution:


1. Describes the number of events occurring in a fixed interval of time or space.
2. Events occur randomly and independently.
3. No upper limit on the number of events that can occur.
4. Discrete distribution.

```python
import numpy as np
sample_size=100
mean=5
# Generate random sample from Poisson distribution
sample=np.random.poisson(mean,sample_size)

# Calculate sample mean and variance
sample_mean=np.mean(sample)
sample_variance=np.var(sample)
print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

**ANS 9**

In both the Binomial and Poisson distributions, the mean (μ) and variance (σ^2) are related as follows:


For Binomial Distribution:


Mean: μ = n * p (n = number of trials, p = probability of success in each trial)


Variance: σ^2 = n * p * (1 - p)


For Poisson Distribution:

Mean: μ = λ (λ = average rate of events)


Variance: σ^2 = λ
