**ANS 1** 
1. Data encoding: Data encoding refers to the process of converting categorical or text-based data into a numerical format that machine learning algorithms can understand.

2. Usefulness: Encoding is crucial as many machine learning algorithms work with numerical data. It allows the incorporation of categorical information into models and ensures the algorithms can interpret and process data effectively.

**ANS 2**

1. Nominal Encoding: It's a technique where categorical data is assigned unique integer labels, preserving the original order or hierarchy.

2. Example: In customer segmentation, if we have a feature like "Education Level" with categories "High School," "College," and "Graduate," we can assign numerical labels like 1, 2, and 3, respectively, maintaining their original order.

**ANS 3**

Preference for Nominal Encoding: Nominal encoding is preferred when the categorical feature has an inherent order or hierarchy that needs to be preserved, like rating scales (low, medium, high).

Example: In survey responses, if the options are "Agree," "Neutral," and "Disagree," nominal encoding retains the ordinality (1, 2, 3), preserving the order of responses.



**ANS 4**

Choice of Encoding Technique: If the categorical data doesn't have an inherent order or hierarchy, one-hot encoding is preferable. It creates binary columns for each unique category, eliminating the possibility of introducing unintended ordinal relationships.

Explanation: One-hot encoding prevents the model from assuming false ordinality or relationships between categories, which could affect the learning process.

**ANS 6**

For categorical features like "species," "habitat," and "diet" without any inherent order or hierarchy, one-hot encoding would be suitable.

Justification: One-hot encoding creates binary columns for each unique category, avoiding the introduction of unintended ordinal relationships and ensuring the machine learning model treats each category equally.

Encoding Techniques:
Gender (Binary): Nominal encoding (assigning 0 or 1 for each gender).

Contract Type: One-hot encoding (creating binary columns for each contract type).



Step-by-step implementation:

Gender Encoding (Nominal): Assigning numerical labels (e.g., 0 for Male, 1 for Female).

Contract Type Encoding (One-Hot): Creating binary columns (e.g., ContractType_Monthly, ContractType_Yearly) and assigning 1 for the corresponding contract type and 0 for others.

This approach converts categorical data into numerical formats suitable for machine learning algorithms, enabling the model to interpret and learn from these features effectively.
