**ANS 1**

1. Filter Method: It's an approach where features are selected based on their statistical properties and correlation with the target variable, independent of the machine learning model.

2. How it works: Filter methods evaluate features using statistical measures like correlation, chi-squared test, mutual information, or variance to rank or score features. Features are then selected based on these scores, regardless of the predictive model used.

**ANS 2**

1. Wrapper Method: Unlike the filter method, wrapper methods involve selecting features based on how well they contribute to the performance of a specific machine learning algorithm. It uses different subsets of features to train the model and evaluates their performance to select the best subset.

**ANS 3**

1. Common Embedded Methods: Techniques like LASSO( Least Absolute Shrinkage and Selection Operator), Ridge Regression, Decision Trees (with feature importance) and Regularized (e.g. Elastic Net) are used for Embedded feature selection.

**ANS 4** 

Drawbacks: Filter methods might ignore interactions between features or fail to consider the impact of feature subsets on model performance. They don't assess the combined effect of features and may select redundant and irrelevant features.


**ANS 5**


Filter Method Usage: The Filter method is preferred when dealing with high-dimensional data, initial feature ranking, or a preliminary understanding of feature importance is required. It's computationally less expensive compared to Wrapper methods.
