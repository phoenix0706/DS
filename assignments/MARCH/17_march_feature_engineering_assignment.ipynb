**ANS 1**

1. Missing Values: Missing values refer to entries in a dataset that are empty or undefined. They can occur due to various reasons like data collection errors, equipment malfunctions, or intentional reasons.


2. Importance of Handling Missing Values: Handling missing values is crucial as they can adversely affect the performance and accuracy of machine learning models, leading to biased results or reduced predictive power.


3. Algorithms Not Affected by Missing Values: Some algorithms like Decision Trees, Random Forests, and XGBoost can handle missing values internally without requiring explicit handling from the user.

**ANS 2**
Techniques to Handle Missing Data:

1. Deletion: Removing rows or columns with missing values.

2. Imputation: Filling missing values with statistics like mean, median, or mode.

3. Prediction: Using machine learning algorithms to predict missing values.

```python
#example -Imputation
import pandas as pd
from sklearn.impute import SimpleImputer

# Creating a DataFrame with missing values
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 6, 7, 8, 9]}
df = pd.DataFrame(data)

# Imputing missing values with mean
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print(df_imputed)
```

**ANS 3**

1. Imbalanced Data: Imbalanced data occurs when the distribution of classes in a data is skewed, with one class significantly outnumbering the others.

2. Impact of Not Handling Imbalanced Data: If not addresed, models trained on imbalanced data tend to favor the majority class, leading to poor performance in predicting the minority class. It might result in biased models that misclassify or ignore the minority class entirely.

**ANS 4** 

1. Up-sampling: Increasing the number of instances in the minority class to balance the dataset. For instance, replicating or creating synthetic examples of the minority class.

2. Down-sampling: Decreasing the number of instances in the majority class to balance the dataset, typically by randomly removing samples.


**ANS 5**

1. Data Augmentation: A technique to artificially increase the size of a dataset by creating modified or new samples from existing data, primarily used in image and text data

2. SMOTE (Synthetic Minority Over-Sampling Technique): A popular method used to up-sample the minority class by generating synthetic samples along the line segments joining similar minority class interfaces.


**ANS 6**

1. Outliers: Outliers are data points that significantly differ from observations in a dataset, often due to measurement errors, data corruption or genuine anomalies.

2. Importance of Handling Outliers: Outliers can skew statistical measures and adversely impact model performance. Handling outliers ensure that the model is not unduly influenced by these extreme values, leading to more robust and accurate predictions
