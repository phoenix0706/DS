**ANS1**:

An exception in python is an error that occurs during the execution of a program and disrupts the normal flow of the program's instructions. Themain difference between exceptions and syntax errors is that syntax errors occur whenthe python interpreter cannot understand the code structure, while exceptions  occur during the execution of the code due to the unforeseen circumstances such as invalid inputs, file not found etc

**ANS2**:

When an exception is not handled in python, it results in the termination of the  program  and displays an error message indicating the type ofexception and the line where it occurred. For example, if we try to divide a number by 0 without handling the "Zero Division Error" the program will terminate with a traceback showing the error.

```python
result = 10 / 0 #program results in ZeroDivisionError  exception
```

**ANS 3**:

try: This block is used to wrap the code that may raise an exception.

except: This block catches and handles exceptions raised in the try block.

else: This block is executed if no exceptions are raised in the try block.

finally: This block is always executed regardless of whether an exception occurs or not, and is typically used for cleanup operations.


```python
try:
    result = 10 / 0
except ZeroDivisionError as e: #exception will be raised
    print("Error:", e)
else:
    print("Division successful!") # since exception is raised else block won't execute
finally:
    print("Cleanup code here") #finally block will always execute
```

```python
try:
    result = 10 / 5
except ZeroDivisionError as e: #no exception raised
    print("Error:", e)
else:
    print("Division successful!") # since no exception is raised else block will execute
finally:
    print("Cleanup code here") #finally block will always execute
```

```python
try:
    # Code block where exceptions may occur
    result = 10 / 2
except ZeroDivisionError as e:
    # Handle the ZeroDivisionError exception
    print("Error:", e)
else:
    # Execute if no exceptions occur in the try block
    print("Division successful!")
finally:
    # Cleanup code that always runs, whether an exception occurred or not
    print("Cleanup code here")
```

```python
class CustomError(Exception):
    pass

def example_function(x):
    if x < 0: 
        raise CustomError("Input cannot be negative")  #raise: is used to raise custom exceptions

try:
    example_function(-5)
except CustomError as e:
    print("Custom Error occurred:", e)
```

**Ans 5**:
Custom exceptions in python are user defined exceptions that inherit from the base "EXCEPTION" class. We need Custom Exceptions to handle specific error conditions unique to our program or application.

By defining custom exception we can provide more descriptive error messages and improve the readability and maintainibility of the code.



```python
class CustomError(Exception):
    pass

def example_function(x):
    if x < 0:
        raise CustomError("Input cannot be negative")

try:
    example_function(-5)
except CustomError as e:
    print("Custom Error occurred:", e)
```
