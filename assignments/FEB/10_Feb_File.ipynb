**ANS 1**
In Python, the open() function is used to open a file.

Different modes of opening a file include:

'r': Open for reading (default).

'w': Open for writing, truncating the file first.

'a': Open for writing, appending to the end of the file if it exists.

'b': Binary mode.

't': Text mode (default).

'+': Open for updating (reading and writing).

Example usage: file = open("filename.txt", "r")


**ANS 2**
The `close()` function is used to close a file after it has been opened. It's important to close a file to release system resources and ensure that all data is written to the file properly.

**ANS 3**


```python
with open("test.txt","w") as file: #if file doesn't exist it will be created
    file.write("This is sample file") # writing to a file
    
with open("test.txt","r")as file: #opening file in read mode
    content=file.read() #reading content
    print(content) 
```

**ANS 4**

`read()` : Reads the entire file content as a single string

`readline()`: Reads a single line from the file.

`readlines()`: Reads all lines to the file and returns them as a list of strings

**ANS 5**

The `with` statement is used with `open()` to automatically close the file when the block is exited.
It's advantage is that it ensures the file is properly closed even if an exception occurs.

**ANS 6**

`write()`: Writes a string to the file.

`writelines()`: Writes a list of strings to the file.

```python
# Writing to a file using write()
with open("output1.txt", "w") as file:
    file.write("Hello, world!\n")

# Writing multiple lines to a file using writelines()
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open("output2.txt", "w") as file:
    file.writelines(lines)
```
