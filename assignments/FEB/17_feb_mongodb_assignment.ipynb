MongoDb is a popular non-relational, document oriented database management system (DBMS). Non relational databases, also known as No SQL databases differ from traditional SQL databases in their data model and storage methods. Rather than storing data in tables with predefined schemes , No SQL databases like MongoDB store data in flexible, JSON like documents. This allows for more dynamic and scalable data storage, making them preffered scenarios where flexibility, scalability and performance are prioritized over strict data consistency and complex transactions. 
Mongodb us often favored over SQL databases in use cases such as web applications, real time analytics, content management systems, and mobile apps where handling large volumes of unstructured or semi structured data efficiency is essential.

**ANS 2**
Features of MongoDB include:
1. Document Oriented: Data is stored in flexible JSON like documents.
2. Schema-Issues: No rigid schema requirements, allowing for easy schema evolution.
3. High Performance: Supports high read and write throughput with horizontal scalability.
4. Rich Query Language: Provides powerful query capabilities including field, range and regex queries.
5. Indexing: Supports efficient indexing for faster query execution.
6. Aggregation Framework: Allows for complex data aggregation operations. 
7. Replication & High Availabiliy:Supports automatic failover and data redundancy through replica sets.
8. Horizontal Scalability: Easily scales out by adding more servers.
9. Geospatial Indexing: Supports geospatial data and queries.
10. Text Search: Provides text search capabililties. 
11. Flexible Deployment: Can be deployed on-premises or in the cloud.
12. Community Support: Large and active community with extensive documentation and resources.

The find() method is used to query a MongoDB database. It returns a cursor object that can be iterated over to retrieve documents matching the query criteria. Here's a simple code to demonstrate the use of find():

```python
# Find documents where age is greater than 30
cursor = coll_create.find({"age": {"$gt": 30}})
for document in cursor:
    print(document)
```

This code retrieves all documents from the collection where the "age" field is greater than 30.

**ANS 6:**

The    `sort()` method in MongoDB is used to sort the documents in a collection based on specified fields. It takes a field or fields to sort by and an optional direction(ascending or descending). 

This below code sorts the documents in the collection by the "age" field in ascending order (from lowest to highest).

```python
cursor = coll_create.find().sort("age", 1)
for document in cursor:
    print(document)
```

1. delete_one(): Deletes a single document that matches the specified filter criteria.
2. delete_many(): Deletes multiple documents that match the specified filter criteria.
3. drop(): Deletes an entire collection from the database.
These methods are used when data needs to be removed from the database. 
delete_one() and delete_many() are used for specific document deletions based on filter criteria, while drop() is used to delete an entire collection.
