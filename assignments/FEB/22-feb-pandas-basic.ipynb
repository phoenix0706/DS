**ANS 3**

```python
# Creating a Pandas DataFrame
data = {'Name': ['Alice', 'Bob', 'Claire'],
        'Age': [25, 30, 27],
        'Gender': ['Female', 'Male', 'Female']}

df = pd.DataFrame(data)

# Printing the DataFrame
print(df)
```

**ANS 4**


DataFrame: A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It can be thought of as a table or spreadsheet with rows and columns. Each column in a DataFrame is a Series. DataFrames allow for handling and manipulating structured data efficiently.


Series: A Series is a one-dimensional labeled array capable of holding any data type. It is essentially a single column of a DataFrame. Series are useful for representing data where the index is meaningful and for performing element-wise operations

```python

# Creating a Series
ages_series = df['Age']

print("DataFrame:")
print(df)
print("\nSeries:")
print(ages_series)
```

**ANS 5**

head(): Returns the first n rows of the DataFrame.

tail(): Returns the last n rows of the DataFrame.

describe(): Generates descriptive statistics of the DataFrame.

info(): Provides a concise summary of the DataFrame, including data 
types and non-null values.
drop(): Removes rows or columns from the DataFrame.

```python
# Using head() to display the first 5 rows of the DataFrame
print(df.head())

# Using describe() to generate descriptive statistics
print(df.describe())
```

```python
# Creating multiple Series
names = pd.Series(['Alice', 'Bob', 'Charlie'])
ages = pd.Series([25, 30, 35])
genders = pd.Series(['Female', 'Male', 'Male'])

# Creating a DataFrame using the Series
df = pd.DataFrame({'Name': names, 'Age': ages, 'Gender': genders})

# Printing the DataFrame
print(df)
```
