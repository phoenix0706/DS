**ANS 1:**

GET: The GET method is used to request data from a specified resource. When a client sends a GET request to a server, it requests data from the server without modifying or updating any resources. GET requests can be cached and bookmarked, and they remain in the browser history. They are typically used for retrieving data from a server.


POST: The POST method is used to submit data to be processed to a specified resource. When a client sends a POST request to a server, it submits data to the server in the request body. POST requests are not cached and cannot be bookmarked, and they do not remain in the browser history. They are typically used for sending form data, uploading files, or making changes to a server's state.


**ANS 2**

The request object in Flask is used to access incoming request data in Flask applications. It allows developers to access form data, file uploads, cookies, and other request information sent by the client. The request object provides methods and attributes to retrieve data from the request, such as request.form to access form data and request.args to access query parameters in a GET request.



**ANS 3**

The redirect() function in Flask is used to redirect the client to a different URL. It is commonly used in web applications to perform URL redirection after processing a request. For example, after a form submission, you might want to redirect the user to a different page to display a success message or to another part of the application. redirect() helps maintain a clean and organized navigation flow in web applications.



**ANS 4**

Templates in Flask are HTML files that contain dynamic content. They allow developers to generate HTML pages dynamically by combining static HTML with dynamic data from the application. Flask uses the Jinja2 templating engine to render templates. The render_template() function in Flask is used to render templates by passing data to be displayed in the template as keyword arguments. It simplifies the process of generating HTML responses in Flask applications.

```
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/api/hello', methods=['GET'])
def hello():
    return jsonify({'message': 'Hello, World!'})

if __name__ == '__main__':
    app.run(debug=True)
```
