**ANS 1**
Flask is a micro web framework for Python that allows developers to build web applications quickly and with minimal overhead. It provides tools, libraries and features to simplify web development tasks, making it a popular choice for building web applications, APIs and more.


Advantages of Flask Framework:

1. Simplicity: Flask has a simple and easy-to-understand syntax, making it  beginner friendly.
2. Lightweight: Flask has minimal dependencies and a small core , resulting in lightweight applications.
3. Flexibility: Flask allows developers to choose components and libraries as needed, providing flexibility in development.
4. Extensibility: Flask can be easily extended with third party extensions and libraries to add additional functionality.
5. Built in Development Server: Flask comes with a built in development server, making it easy to test and debug applications.
6. Jinja2 Templating: Flask uses Jinja2 templating engine, which provides powerful template inheritance and macros for building dynamic HTML pages.
