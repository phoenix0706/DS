**ANS 1**:

A database is a structured collection of data that is organized in such a way that it can be easily accessed, managed and updated . It can be thought of as an electronic filing system.

SQL (Structured Query Language) databases are relational databases that use a structured query language for defining and manipulating data. They are based on the relational model and use tables to store data. Examples include MySql, Postgresql, and Oracle.


NoSQL (Not only SQL) databases are non relational databases that provide a mechanism for storage and retrieval of data that is modeled in means other than tabular relations used in relational databases. They are often used forlarge sets of distributed data. Examples include Mongodb, Cassandra and Redis.

**ANS 2**

DDL (Data Definition Language) is a subset of SQL used for defining, altering, and dropping database objects such as tables, indexes, and views.


CREATE: Used to create new database objects. For example, CREATE TABLE is used to create a new table.


DROP: Used to delete database objects. For example, DROP TABLE is used to delete a table.


ALTER: Used to modify database objects. For example, ALTER TABLE is used to add or remove columns from a table.


TRUNCATE: Used to remove all records from a table without deleting the table itself. For example, TRUNCATE TABLE is used to remove all rows from a table.

**ANS 3**

DML (Data Manipulation Language) is a subset of SQL used for manipulating data within database objects.



INSERT: Used to insert new records into a table. For example, INSERT INTO table_name (column1, column2) VALUES (value1, value2);



UPDATE: Used to modify existing records in a table. For example, UPDATE table_name SET column1 = value1 WHERE condition;



DELETE: Used to remove records from a table. For example, DELETE FROM table_name WHERE condition;

**ANS 4** 


DQL (Data Query Language) is a subset of SQL used for querying data from database objects.

SELECT: Used to retrieve data from one or more tables. For example, SELECT column1, column2 FROM table_name WHERE condition;


**ANS 5**

Primary Key: A primary key is a column or a set of columns that uniquely identifies each row in a table. It must contain unique values and cannot contain null values. Example: id column in a users table.


Foreign Key: A foreign key is a column or a set of columns in one table that refers to the primary key in another table. It establishes a link between the two tables. Example: user_id column in an orders table referencing the id column in the users table.

cursor(): This method creates a cursor object which is used to execute SQL queries.

execute(): This method is used to execute SQL queries. You pass the SQL query as a parameter to this method.

```python



import mysql.connector

# Establish connection
mydb = mysql.connector.connect(
  host="localhost",
  user="yourusername",
  password="yourpassword",
  database="yourdatabase"
)

# Create cursor object
mycursor = mydb.cursor()

# Execute SQL query
mycursor.execute("SELECT * FROM yourtable")

# Fetch result
result = mycursor.fetchall()

# Print result
for row in result:
  print(row)
```
