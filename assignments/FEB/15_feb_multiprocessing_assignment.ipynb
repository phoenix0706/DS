**ANS 1**

Multiprocessing in Python refers to the technique of using multiple processes execute tasks concurrently. It allows you to take advantage of multiple CPU cores available on your system, which can lead to improved performance for CPU-bound tasks. Each process has its own memory and space and resources, making it possible to perform parallel processing in Python.

**ANS 2**


1. Multiprocessing involves multiple process running concurrenlty, with each process having its own memory space and resources.
2. Multithreading involves multiple threads running within a single process, sharing the same memory space and resources.
3. Multiprocessing is more suitable for CPU-bound tasks, as it can take advantage of multiple cores.
4. Multithreading is more suitable for I/O bound tasks, where threads can work on different tasks while waiting for I/O operations to complete


**ANS 3**

```python
import multiprocessing
def worker_function():
    print('This is a worker process.')
if __name__=='__main__':
    process=multiprocessing.Process(target=worker_function)
    process.start()
    process.join()
    print('Main process completed')
```

**ANS 4**

A multiprocessing pool is a way to manage a group of worker processes that can be used to parallize the execution of a function across multiple input values. It allows you to submit multiple tasks to the pool, and the pool automatically distributes those tasks among the available worker processes. This is particularly useful when you want to apply the same function to a large set of input data concurrently

**ANS 5**

```python
import multiprocessing
def worker_function(x):
    return x*x
if __name__=='__main__':
    with multiprocessing.Pool(processes=4) as pool:
        results =pool.map(worker_function, range(10))
    print('Results:',results)
```

**ANS 6**

```python
import multiprocessing

def worker_function(number):
    print("Process", number, "printed", number)

if __name__ == "__main__":
    processes = []
    for i in range(4):
        process = multiprocessing.Process(target=worker_function, args=(i,))
        processes.append(process)
        process.start()

    for process in processes:
        process.join()

    print("Main process completed.")
```
