```python
def flatten_and_calculate_product(input_list):
    flat_list=[]
    def flatten(lst):
        for item in lst:
        
            if isinstance(item, (list,tuple,set)):
                
                flatten(item)
            else:
                flat_list.append(item)
    flatten(input_list)
    product=1
    for item in flat_list:
        if isinstance(item,(int,float)) and not isinstance(item,(bool)):
            product*=item
    return product

# Test data
list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']
result = flatten_and_calculate_product(list1)
print("Product:", result)

```

```python
def encrypt_message(message):
    encrypted=""
    for char in message:
        if char.isalpha():
            encrypted_char = chr(ord('a') + (ord('z') - ord(char)))
            encrypted += encrypted_char
        elif char ==' ':
            encrypted+='$'
        else:
            encrypted+=char
    return encrypted

# Input sentence
input_sentence = "I want to become a Data Scientist."
lowercase_input = input_sentence.lower()  # Convert to lowercase
encrypted_output = encrypt_message(lowercase_input)
print("Encrypted:", encrypted_output)
```
